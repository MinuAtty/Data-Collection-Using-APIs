```python
# Importing the requests library to make HTTP requests
import requests
```

```python
# Defining the base URL for the Pokémon API
base_url = 'https://pokeapi.co/api/v2/'
```

```python
# Function to get Pokémon information from the API based on its name
def get_pokemon_info(name):
    # Construct the URL for the specific Pokémon by appending its name to the base URL
    url = f'{base_url}pokemon/{name}'

    # Sending a GET request to the constructed URL
    response = requests.get(url)

    # Checking if the request was successful (status code 200 means OK)
    if response.status_code == 200:
        # Parsing the response JSON into a Python dictionary
        pokemon_data = response.json()
        # Returning the parsed Pokémon data
        return pokemon_data
    else:
        # If the request was not successful, print an error message and return None
        print(f"Failed to retrieve data {response.status_code}")

# Specifying the name of the Pokémon to query
pokemon_name = 'sylveon'

# Calling the function to fetch data for the specified Pokémon and storing the result
pokemon_info = get_pokemon_info(pokemon_name)
```

```python
# If the Pokémon data was successfully retrieved
if pokemon_info:
    print(f"Name: {pokemon_info['name'].capitalize()}")   # Printing the Pokémon's name, capitalized for better readability
    print(f"ID: {pokemon_info['id']}")                    # Printing the Pokémon's ID
    print(f"Height: {pokemon_info['height']}")            # Printing the Pokémon's height (as provided by the API)
    print(f"Weight: {pokemon_info['weight']}")            # Printing the Pokémon's weight (as provided by the API)
    print(f"Abilities: {', '.join(ability['ability']['name'] for ability in pokemon_info['abilities'])}")         # Extracting and printing a comma-separated list of the Pokémon's abilities
    print(f"Types: {', '.join(type['type']['name'] for type in pokemon_info['types'])}")                          # Extracting and printing a comma-separated list of the Pokémon's types
```
